```

X_train = x[0:600,:28]
X_test = x[600:740,:28]

print(len(X_train))
print(len(X_train[0]))
print("-------")
print(len(X_test))
print(len(X_test[0]))


X2_train = kol_sy[0:600,:]
X2_test = kol_sy[600:740,:]

print(len(X2_train))
print(len(X2_train[0]))
print("******************")
print(len(X2_test))
print(len(X2_test[0]))


```

```
Y_train = kol_sy[0:600,:27]
Y_test = kol_sy[600:740,:27]

Y2_train=out_in[0:600,:27]
Y2_test=out_in[600:740,:27]

print(len(Y_train))
print(len(Y_train[0]))
print("-------")
print(len(Y_test))
print(len(Y_test[0]))
num_classes=8
X_train=X_train.reshape(X_train.shape[0],27).astype('float32')
X_test=X_test.reshape(X_test.shape[0],27).astype('float32')


print(len(Y2_train))
print(len(Y2_train[0]))
print("-------")
print(len(Y2_test))
print(len(Y2_test[0]))


```

```
def baseline_model():
  model=Sequential()
  model.add(Dense(740,input_dim=27,activation='relu'))
  model.add(Dense(15,activation='sigmoid'))
  model.add(Dropout(0.2))
  model.add(Dense(40,activation='softmax'))
  model.add(Dropout(0.1))
  model.add(Dense(1,activation='relu'))
  


  model.compile(loss='mean_squared_error',optimizer='SGD',metrics=['accuracy'])

  return model


def Secondbaseline_model():
  model=Sequential()
  model.add(Dense(50,input_dim=1,activation='relu'))
  model.add(Dense(15,activation='softmax'))
  model.add(Dense(20,activation='relu'))
  model.add(Dropout(0.1))
  model.add(Dense(40,activation='softmax'))
  model.add(Dense(1,activation='relu'))

  model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

  return model
```

```
model=baseline_model()
history=model.fit(X_train,Y_train,epochs=2000,batch_size=200)
scores=model.evaluate(X_test,Y_test)
print("Base line error: %.2f%%" %(100-scores[1]*100))

predictions = model.predict(X_test)
error = abs(Y_test - predictions)

print(history.history['loss'])
print(np.average(error))

```

```
models=Secondbaseline_model()
y2=models.fit(X2_train,Y2_train,epochs=600,batch_size=100)
y2_scores=models.evaluate(X2_test,Y2_test)
print("SecondBase line error: %.2f%%" %(100-y2_scores[1]*100))
print('Accuracy: %.2f' % (y2_scores[1]*100))
```
